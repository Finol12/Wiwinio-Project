## Testing queries on vivino.db with SQLite3.
Getting a basic query set-up for Wiwinio Project.

```python
import sqlite3

connection = sqlite3.connect('../data/vivino.db')
cursor = connection.cursor()

cursor.execute ('''
    SELECT
        vintages.name,
        ROUND(vintages.price_euros, 2) AS nrs
    FROM
        wines
        JOIN vintages ON vintages.wine_id = wines.id
    WHERE
        vintages.price_euros <= 200
    GROUP BY
        price_euros
    ORDER BY
        nrs DESC;
''')
cursor.fetchall()



```

```python
cursor.execute ('''
    SELECT
        countries.name,
        vintages.name,
        regions.name,
        ROUND(vintages.price_euros, 2) AS price
    FROM
        countries
        JOIN regions ON countries.code = regions.country_code
        JOIN wines ON regions.id = wines.region_id
        JOIN vintages on wines.id = vintages.wine_id
    WHERE
        vintages.price_euros >= 10000
    GROUP BY
        price
    ORDER BY
        price DESC;
''')
cursor.fetchall()
```

```python
cursor.execute ('''
    SELECT
        countries.name,
        vintages.name,
        regions.name,
        toplists.name,
        vintage_toplists_rankings.rank,
        ROUND(vintages.price_euros, 2) AS price
    FROM
        countries
        JOIN regions ON countries.code = regions.country_code
        JOIN toplists ON countries.code = toplists.country_code
        JOIN vintage_toplists_rankings on toplists.id = vintage_toplists_rankings.top_list_id
        JOIN wines ON regions.id = wines.region_id
        JOIN vintages on wines.id = vintages.wine_id
        
    WHERE
        vintages.price_euros <= 1000
    GROUP BY
        price
    ORDER BY
        price DESC;
''')
cursor.fetchall()
```

```python
cursor.execute ('''
    SELECT
        countries.name AS country,
        ROUND(AVG(wines.ratings_average), 2) AS hra
    FROM
        countries
        JOIN regions ON countries.code = regions.country_code
        --JOIN toplists ON countries.code = toplists.country_code
        --JOIN vintage_toplists_rankings on toplists.id = vintage_toplists_rankings.top_list_id
        JOIN wines ON regions.id = wines.region_id
        JOIN vintages on wines.id = vintages.wine_id
    GROUP BY
        country
    ORDER BY
        hra DESC;
''')
cursor.fetchall()
```

```python
cursor.execute ('''
    SELECT
        vintages.year AS year,
        ROUND(AVG(wines.ratings_average), 2) AS hra
    FROM
        countries
        JOIN regions ON countries.code = regions.country_code
        JOIN toplists ON countries.code = toplists.country_code
        JOIN vintage_toplists_rankings on toplists.id = vintage_toplists_rankings.top_list_id
        JOIN wines ON regions.id = wines.region_id
        JOIN vintages on wines.id = vintages.wine_id
    GROUP BY
        year
    ORDER BY
        hra DESC
    LIMIT 10;
''')
cursor.fetchall()
```
